# **A Simple Tutorial on how to open a Virtual Environment**


Virtual Environment: (shortened as "virtualenv") separate our new project's Python dependencies from our other projects and from the Python libraries our operating system uses. If you don't use a virtualenv, there's a good chance you might break part of your OS.

**On the terminal that you're running your code, put in these commands:**

>python3 -m venv venv

>source venv/bin/activate

(venv)(base) will show, now write:
> pip3 install flask

**When to use venv?**

Usually used when using Flask.

Here is a simple flask application to test your virtual env:

```python
from flask import Flask

app = Flask(__name__)

@app.route("/") #for homepage
def hello_world():
    return "Hello, World"
```
